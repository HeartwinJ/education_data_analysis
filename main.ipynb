```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_all_data(df):
	with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.precision', 3):
		display(df)
```

```python
df = pd.read_csv('global_education_data.csv')
df.head()
```

```python
print(f"Columns in dataset:\n {df.columns}")
print(f"Shape of dataset:\n {df.shape}")
print(f"Number of rows in dataset:\n {df.shape[0]}")
```

```python
df.drop(columns=['Latitude', 'Longitude', 'Grade_2_3_Proficiency_Reading',
			 'Grade_2_3_Proficiency_Math', 'Primary_End_Proficiency_Reading',
			 'Primary_End_Proficiency_Math', 'Lower_Secondary_End_Proficiency_Reading',
       'Lower_Secondary_End_Proficiency_Math', 'Youth_15_24_Literacy_Rate_Male',
			 'Youth_15_24_Literacy_Rate_Female', 'Birth_Rate', 'Gross_Primary_Education_Enrollment',
       'Gross_Tertiary_Education_Enrollment'], inplace=True)
df.head()
```

```python
print(f"Columns in dataset:\n {df.columns}")
print(f"Shape of dataset:\n {df.shape}")
print(f"Number of rows in dataset:\n {df.shape[0]}")
```

## Objectives

### Aim 1: Analyze out of school rates

#### 1.1 Top 10 Countries with the highest Out of School Rates

```python
cols = ['OOSR_Pre0Primary_Age_Male', 'OOSR_Pre0Primary_Age_Female',
				'OOSR_Primary_Age_Male', 'OOSR_Primary_Age_Female',
				'OOSR_Lower_Secondary_Age_Male', 'OOSR_Lower_Secondary_Age_Female',
				'OOSR_Upper_Secondary_Age_Male', 'OOSR_Upper_Secondary_Age_Female']
oosr_aggr = sum([df[col] for col in cols]) / len(cols)

df1_1 = pd.DataFrame({'Country': df['Country'], 'OOSR': oosr_aggr})
df1_1.head()
```

```python
largest10 = df1_1.nlargest(10, 'OOSR')
print_all_data(largest10)
largest10.plot(x='Country', y='OOSR', ylim=(0, 100), kind='bar', rot=90)
```

### Aim 3: Investigate relationship between education indicators and unemployment rates

#### 3.1 Top 10 Countries with the highest Unemployment rates

```python
highest_unemployment = df[['Country', 'Unemployment_Rate']].nlargest(10, 'Unemployment_Rate')
print_all_data(highest_unemployment)
highest_unemployment.plot(x='Country', y='Unemployment_Rate', kind='bar', rot=90)
```

#### 3.2 Top 10 Countries with the lowest Unemployment rates

```python
lowest_unemployment = df[df['Unemployment_Rate'] != 0][['Country', 'Unemployment_Rate']].nsmallest(10, 'Unemployment_Rate')
print_all_data(lowest_unemployment)
lowest_unemployment.plot(x='Country', y='Unemployment_Rate', kind='bar', rot=90)
```

#### 3.3 Relation between Unemployment and Completion and Out of School Rates

##### 3.3.1 Unemployment vs Completion

```python
unemployment_per_country = df['Unemployment_Rate']
completion_cols = ['Completion_Rate_Primary_Male',
									 'Completion_Rate_Primary_Female',
									 'Completion_Rate_Lower_Secondary_Male',
									 'Completion_Rate_Lower_Secondary_Female',
									 'Completion_Rate_Upper_Secondary_Male',
									 'Completion_Rate_Upper_Secondary_Female']
completion_aggr = sum([df[col] for col in completion_cols]) / len(completion_cols)

df3_3_1 = pd.DataFrame({'Country': df['Country'], 'Unemployment_Rate': unemployment_per_country, 'Completion_Rate': completion_aggr})

df3_3_1.plot(x='Country', y=['Unemployment_Rate', 'Completion_Rate'], ylim=(0,100), kind='bar', rot=90, figsize=(50, 10))
```

##### 3.3.2 Unemployment vs Out of School Rate

```python
unemployment_per_country = df['Unemployment_Rate']
oosr_cols = ['OOSR_Pre0Primary_Age_Male',
						 'OOSR_Pre0Primary_Age_Female', 'OOSR_Primary_Age_Male',
						 'OOSR_Primary_Age_Female', 'OOSR_Lower_Secondary_Age_Male',
						 'OOSR_Lower_Secondary_Age_Female', 'OOSR_Upper_Secondary_Age_Male',
						 'OOSR_Upper_Secondary_Age_Female']
oosr_aggr = sum([df[col] for col in oosr_cols]) / len(oosr_cols)

df3_3_1 = pd.DataFrame({'Country': df['Country'], 'Unemployment_Rate': unemployment_per_country, 'OOSR_Rate': oosr_aggr})

df3_3_1.plot(x='Country', y=['Unemployment_Rate', 'OOSR_Rate'], ylim=(0,100), kind='bar', rot=90, figsize=(50, 10))
```
